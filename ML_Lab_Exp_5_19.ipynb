```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import pandas as pd
import numpy as np
```

```python
data = load_breast_cancer()

x = data.data
y = data.target

print("Classes: ", data.target_names)
```

```python
df = pd.DataFrame(x)
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
lr = LogisticRegression(max_iter=5000)

lr.fit(X_train, y_train)

pred_lr = lr.predict(X_test)

print("Logistic Test Accuracy:", accuracy_score(y_test, pred_lr))

pred_lr_train = lr.predict(X_train)

print("Logistic Train Accuracy:", accuracy_score(y_train, pred_lr_train))
```

```python
dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)

pred_dt = dt.predict(X_test)

print("DecisionTree Test Accuracy:", accuracy_score(y_test, pred_dt))

pred_dt_train = dt.predict(X_train)

print("DecisionTree Train Accuracy:", accuracy_score(y_train, pred_dt_train))
```

```python
knn = KNeighborsClassifier()

knn.fit(X_train, y_train)

pred_knn = knn.predict(X_test)

print(" KNeighbors Test Accuracy:", accuracy_score(y_test, pred_knn))

pred_knn_train = knn.predict(X_train)

print(" KNeighbors Train Accuracy:", accuracy_score(y_train, pred_knn_train))
```

```python
gnb = GaussianNB()

gnb.fit(X_train, y_train)

pred_gnb = gnb.predict(X_test)

print(" GaussianNB Test Accuracy:", accuracy_score(y_test, pred_gnb))

pred_gnb_train = gnb.predict(X_train)

print(" GaussianNB Train Accuracy:", accuracy_score(y_train, pred_gnb_train))
```

```python
mlp = MLPClassifier()

mlp.fit(X_train, y_train)

pred_mlp = mlp.predict(X_test)

print(" MLPClassifier Test Accuracy:", accuracy_score(y_test, pred_mlp))

pred_mlp_train = mlp.predict(X_train)

print(" MLPClassifier Train Accuracy:", accuracy_score(y_train, pred_mlp_train))
```

```python
cm_lr = confusion_matrix(y_test, pred_lr)
print("\nLinear Regression Confusion Matrix")
print(cm_lr)
TN1, FP1, FN1, TP1 = cm_lr.ravel()

type1_lr = FP1 / (FP1 + TN1) 
type2_lr = FN1 / (FN1 + TP1) 

print("FP (Type-I): ",FP1)
print("FN (Type-II): ",FN1)
print("Type-I error rate in %:", type1_lr*100)
print("Type-II error rate in %:", type2_lr*100)
```

```python
cm_dt = confusion_matrix(y_test, pred_dt)

print("\nDecision Tree Confusion Matrix:")
print(cm_dt)

TN2, FP2, FN2, TP2 = cm_dt.ravel()

type1_dt = FP2 / (FP2 + TN2)
type2_dt = FN2 / (FN2 + TP2)

print("FP (Type-I):", FP2)
print("FN (Type-II):", FN2)
print("Type-I Error Rate in %:", type1_dt * 100)
print("Type-II Error Rate in %:", type2_dt * 100)
```

```python
cm_knn = confusion_matrix(y_test, pred_knn)
print("\nKNN Confusion Matrix")
print(cm_knn)
TN3, FP3, FN3, TP3 = cm_knn.ravel()

type1_knn = FP3 / (FP3 + TN3) 
type2_knn = FN3 / (FN3 + TP3) 

print("FP (Type-I): ",FP3)
print("FN (Type-II): ",FN3)
print("Type-I error rate in %:", type1_knn*100)
print("Type-II error rate in %:", type2_knn*100)
```

```python
cm_nb = confusion_matrix(y_test, pred_gnb)
print("\nNaive Bayes Confusion Matrix")
print(cm_nb)
TN4, FP4, FN4, TP4 = cm_nb.ravel()

type1_nb = FP4 / (FP4 + TN4) # False positive rate
type2_nb = FN4 / (FN4 + TP4) # False negative rate

print("FP (Type-I): ",FP4)
print("FN (Type-II): ",FN4)
print("Type-I error rate in %:", type1_nb*100)
print("Type-II error rate in %:", type2_nb*100)
```

```python
cm_mlp = confusion_matrix(y_test, pred_mlp)
print("\nNeural Network Confusion Matrix")
print(cm_mlp)
TN5, FP5, FN5, TP5 = cm_mlp.ravel()

type1_mlp = FP5 / (FP5 + TN5) # False positive rate
type2_mlp = FN5 / (FN5 + TP5) # False negative rate

print("FP (Type-I): ",FP5)
print("FN (Type-II): ",FN5)
print("Type-I error rate in %:", type1_mlp*100)
print("Type-II error rate in %:", type2_mlp*100)
```

```python
precision = precision_score(y_test, pred_lr)
recall = recall_score(y_test, pred_lr)
f1 = f1_score(y_test, pred_lr)
roc = roc_auc_score(y_test, pred_lr)

print("precision:", precision)
print("recall:", recall)
print("f1:", f1)
print("roc:", roc)
```
